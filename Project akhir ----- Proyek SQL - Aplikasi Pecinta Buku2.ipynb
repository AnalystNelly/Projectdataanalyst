## Tugas

### Temukan jumlah buku yang dirilis setelah 1 Januari 2000

```python
# Temukan jumlah buku yang dirilis setelah 1 Januari 2000.

# Membuat query SQL
query_satu = """
SELECT count(*) AS jumlah_buku
FROM public.books
WHERE publication_date > '2000-01-01';
"""

# Membaca hasil query ke dalam dataframe
baca_query = pd.io.sql.read_sql(query_satu, con=engine)

# Menampilkan hasil
print(baca_query)
```

### Hitung jumlah ulasan pengguna dan rating rata-rata untuk setiap buku

```python
# Tugas :
# ● Temukan jumlah ulasan pengguna dan rata-rata peringkat untuk setiap buku.  

# Membuat query ke-dua
query_dua = """
SELECT 
    b.book_id, 
    b.title AS judul_buku, 
    COUNT(r.review_id) AS jumlah_ulasan,
    AVG(rt.rating) AS rata_rata_peringkat 
FROM public.books as b
LEFT JOIN 
    public.reviews AS r ON b.book_id = r.book_id
LEFT JOIN 
    public.ratings AS rt ON b.book_id = rt.book_id
GROUP BY 
    b.book_id, title
ORDER BY
    jumlah_ulasan;
"""

# Membaca hasil kueri ke dalam dataframe
baca_query = pd.io.sql.read_sql(query_dua, con=engine)

# Menampilkan hasil
#print(df)
baca_query

```

Kesimpulan: 

Kita telah menghitung jumlah ulasan pengguna dan rata-rata peringkat untuk setiap buku. 
Informasi ini dapat memberikan wawasan tentang popularitas dan kualitas buku-buku tertentu, yang dapat digunakan dalam pengambilan keputusan terkait dengan perpustakaan atau penjualan buku.

### Identifikasi penerbit yang telah menerbitkan jumlah buku terbanyak

```python
#tugas :
#● Identifikasi penerbit yang telah merilis jumlah buku terbanyak dengan lebih dari 50 halaman 
#  (ini akan membantu Anda mengeluarkan brosur dan publikasi serupa dari analisis Anda).

# Membuat query ke-tiga
query_tiga = """
SELECT
    p.publisher_id,
    p.publisher AS nama_penerbit,
    COUNT(b.book_id) AS jumlah_buku_terbit
FROM
    public.publishers AS p
JOIN
    public.books AS b ON p.publisher_id = b.publisher_id
WHERE
    num_pages > 50
GROUP BY
    p.publisher_id, nama_penerbit
ORDER BY
    jumlah_buku_terbit DESC
LIMIT 10;
"""

# Membaca  hasil query kedalam dataframe
baca_query = pd.io.sql.read_sql(query_tiga, con=engine)

# Menampilkan hasil 
#print(baca_query)
baca_query
```

Kesimpulan: 
Data yang di tampilkan 10 raw teratas.

[Penguin Books] adalah penerbit yang telah merilis jumlah buku terbanyak dengan lebih dari 50 halaman. Informasi ini dapat membantu dalam mengenali penerbit yang memiliki kontribusi besar dalam industri buku dalam menerbitkan buku-buku.

### Identifikasi penulis dengan rata-rata peringkat buku tertinggi

```python
# ● Identifikasi penulis dengan rata-rata peringkat buku tertinggi (lihat hanya buku dengan setidaknya 50 ulasan).

# Membuat query ke-empat
query_empat = """
SELECT
    a.author_id,
    a.author AS nama_penulis,
    AVG(rt.rating) AS rata_rata_peringkat,
    COUNT(*) AS jumlah_buku
FROM
    public.authors AS a
JOIN
    public.books AS b ON a.author_id = b.author_id
JOIN
    (
        SELECT book_id
        FROM public.ratings
        GROUP BY book_id
        HAVING COUNT(*) >= 50
    ) AS rated_books ON b.book_id = rated_books.book_id
JOIN
    public.ratings AS rt ON b.book_id = rt.book_id
GROUP BY
    a.author_id, nama_penulis
ORDER BY
    rata_rata_peringkat DESC
LIMIT 1;
"""

# Membaca hasil kueri ke dalam dataframe
baca_query = pd.io.sql.read_sql(query_empat, con=engine)

# Menampilkan hasil
baca_query


```

Kesimpulan: 

Data yang ditampilkan tidak ada limitnya.

[J.K. Rowling/Mary GrandPré] adalah penulis dengan rata-rata peringkat buku tertinggi, memiliki rata-rata_peringkat [4.287097] dengan jumlah buku [310] ini adalah data bahwa hanya buku dengan setidaknya 50 ulasan yang dipertimbangkan. 

Informasi ini dapat digunakan untuk mengetahui kualitas tulisan penulis dalam menghasilkan buku-buku yang mendapatkan peringkat tinggi dari pembaca.

### Temukan rata-rata jumlah ulasan teks di antara pengguna yang memberi peringkat lebih dari 50 buku

```python
#● Temukan rata-rata jumlah ulasan teks di antara pengguna yang memberi peringkat lebih dari 50 buku.

# membuat query_lima
query_lima = """
SELECT AVG(count) AS rata_rata_jumlah_ulasan_teks
FROM (
    SELECT COUNT(reviews.text) 
    FROM
        (SELECT username
        FROM ratings
        GROUP BY username
        HAVING COUNT(rating) > 50 )AS users
    JOIN reviews ON users.username = reviews.username
    GROUP BY users.username) AS count
    
"""

# Membaca hasil kueri ke dalam dataframe
baca_query = pd.io.sql.read_sql(query_lima, con=engine)

# Menampilkan hasil
#print(baca_query)
baca_query
```

Kesimpulan: 

Rata-rata jumlah ulasan teks di antara pengguna yang memberi peringkat lebih dari 50 buku adalah [24.33]. 

Ini menggambarkan tingkat partisipasi dan keterlibatan pengguna dalam memberikan ulasan teks, 
Hal ini dapat berguna dalam memahami tingkat interaksi antara pengguna dengan buku-buku yang telah diterbitkan.
