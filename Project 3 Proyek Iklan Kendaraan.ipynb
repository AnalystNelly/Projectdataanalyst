## Memperbaiki Kualitas Data

[Berikan faktor tambahan pada data agar lebih mudah dianalisis.]

```python
# Tambahkan nilai waktu dan tanggal pada saat iklan ditayangkan 
data_iklan['datetime_posted'] = pd.to_datetime(data_iklan['date_posted'])
data_iklan
```

```python
# Tambahkan usia kendaraan saat iklan ditayangkan
current_year = pd.Timestamp.now().year
data_iklan['car_age'] = current_year - data_iklan['model_year']
data_iklan
```

```python
# Tambahkan jarak tempuh rata-rata kendaraan per tahun 
data_iklan['avg_per_year'] = data_iklan['odometer'] / data_iklan['car_age']
data_iklan
```

```python
# Mungkin membantu untuk mengganti nilai pada kolom 'condition' dengan sesuatu yang dapat dimanipulasi dengan lebih mudah 
data_iklan['condition'] = data_iklan['condition'].replace({
    'salvage': 0,
    'fair': 1,
    'good': 2,
    'excellent': 3,
    'like new': 4,
    'new': 5
})
data_iklan
```

```python
# penggunaan fungsi untuk memplot histogram dan boxplot guna menghindari penduplikasian kode

import matplotlib.pyplot as plt

def plot_histo_boxplot(df, col_name):
    fig, ax = plt.subplots(1, 2, figsize=(12, 5))
    
    # plot histogram
    df[col_name].plot(kind='hist', bins=50, ax=ax[0])
    ax[0].set_title('Histogram')
    
    # plot boxplot
    df.boxplot(column=col_name, ax=ax[1], vert=False)
    ax[1].set_title('Boxplot')
    
    plt.show()
```

```python
# plot histogram and boxplot berdasarkan price
plot_histo_boxplot(data_iklan, 'price')
```

## Mempelajari dan Menangani Outlier

[Dengan hasil di atas, tentukan kolom apa yang mungkin berisi *outlier*, dan buatlah sebuah DataFrame tanpa *outlier* tersebut. Petunjuk: *Outlier* dapat diidentifikasi dengan menentukan batas bawah/atas untuk rentang nilai normal.]

```python
data_iklan.describe()
```

```python
# Tentukan batas atas - bawah outlier 

#menangani outlier pada kolom 'price
Q1 = data_iklan['price'].quantile(0.25)
Q3 = data_iklan['price'].quantile(0.75)

IQR = Q3 - Q1

# Tentukan batas bawah outlier 
lower_price = Q1 - 1.5*IQR

# Tentukan batas atas outlier 
upper_price = Q3 + 1.5*IQR

#data_iklan_no_outlier = data_iklan[(data_iklan['price'] >= lower_price) & (data_iklan['price'] <= upper_price)]
no_outlier_price = data_iklan.drop(data_iklan[(data_iklan['price'] < lower_price) | (data_iklan['price'] > upper_price)].index)


print("Batas bawah harga :", lower_price) 
print("Batas atas harga :", upper_price) 

```

```python
#menangani outlier pada kolom 'model_year'
Q1 = data_iklan['model_year'].quantile(0.25)
Q3 = data_iklan['model_year'].quantile(0.75)

IQR = Q3 - Q1

# Tentukan batas bawah outlier 
lower_model_year = Q1 - 1.5*IQR

# Tentukan batas atas outlier 
upper_model_year = Q3 + 1.5*IQR

#data_iklan_no_outlier = data_iklan[(data_iklan['model_year'] >= lower_year) & (data_iklan['model_year'] <= upper_year)]
no_outlier_year = data_iklan.drop(data_iklan[(data_iklan['model_year'] < lower_model_year) | (data_iklan['model_year'] > upper_model_year)].index)

print("Batas bawah Tahun kendaraan :", lower_model_year) 
print("Batas atas tahun kendaraan :", upper_model_year) 
```
