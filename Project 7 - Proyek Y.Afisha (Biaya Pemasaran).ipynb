####  4. Berapa jumlah uang yang mereka kontribusikan? (LTV)

```python
data_orders['buy_ts'] = pd.to_datetime(data_orders['buy_ts'])

#data_orders['orders_month'] = data_orders['orders_month'].astype('datetime64[M]')
```

```python
data_orders.head() 
```

```python
# Memukan jumlah pengguna unik di setiap kohort.
# Memukan bulan dilakukannya pesanan pertama oleh setiap pelanggan

first_orders = data_orders.groupby('uid').agg({'first_orders_month': 'min'}).reset_index()
#first_orders.columns = ['uid','first_orders_month']
first_orders.head() 
```

```python
#Menghitung jumlah pelanggan yang membuat pesanan pertama tiap bulan
cohort_sizes = first_orders.groupby('first_orders_month').agg({'uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_orders_month', 'n_buyers']
print(cohort_sizes.head())
```

```python
# Menambahkan bulan dilakukannya pesanan pertama ke DataFrame pembelian
orders_first = pd.merge(data_orders, first_orders, on='uid')
orders_first
```

```python
#Mengelompokkan tabel "data_orders" berdasarkan bulan pembelian pertama dan bulan pembelian.
cohorts = data_orders.groupby(['first_orders_month','orders_month']).agg({'revenue': 'sum'}).reset_index()
cohorts.head() 
```

```python
# kelompokkan pesanan ke dalam kohort
cohorts = (data_orders.groupby(['first_orders_month', 'orders_month']).agg({'revenue': 'sum'}).reset_index())
```

```python
#Menyusun laporan kohort dengan menambahkan kolom usia kohort dan total laba kotor untuk setiap kohort per bulannya
reports = pd.merge(cohort_sizes, cohorts, on='first_orders_month')
reports.head()
```

```python
# Kita tentukan margin laba = 0.4
margin_rate = 0.4

# hitung laba kotor
reports['laba_kotor'] = reports['revenue'] * margin_rate

# temukan usia setiap kohort dalam satuan bulan, pastikan nilainya berupa integer/bilangan bulat
reports['age'] = (reports['orders_month'] - reports['first_orders_month']) / np.timedelta64(1, 'M')
reports['age'] = reports['age'].round().astype('int')

# hitung LTV untuk cohorts
reports['ltv'] = reports['laba_kotor'] / reports['n_buyers']

```

```python
# Menampilkan hasil reports :
reports.head()
```

```python
# # hitung nilai untuk pivot table
results = reports.pivot_table(index='first_orders_month', columns='age', values='ltv', aggfunc='mean').round()

results = results.fillna('')

results
```

```python
#Mari kita cari LTV dari kohort pertama dengan menjumlahkan nilai untuk setiap bulan:
ltv_201706 = results.loc['2017-06-01'].sum()
ltv_201706 
```

```python
data_costs['month'] = data_costs['dt'].astype('datetime64[M]')
```

```python
# hitung biaya bulanan
monthly_costs = data_costs.groupby('month')['costs'].sum()
monthly_costs
```

```python
# tambahkan biaya bulanan ke laporan kohort
reports_cohort = pd.merge(reports,monthly_costs, left_on='first_orders_month', right_on='month')
```

```python
# hitung CAC
reports_cohort['cac'] = reports_cohort['costs'] / reports_cohort['n_buyers']
reports_cohort
```

Selanjutnya, mari kita cari ROMI (return on marketing investment atau pengembalian investasi pemasaran) dengan membagi LTV dengan CAC.

```python
# hitung ROMI
reports_cohort['romi'] = reports_cohort['ltv'] / reports_cohort['cac']

# buat pivot table baru
output = reports_cohort.pivot_table(index='first_orders_month', columns='age',values='romi', aggfunc='mean')
```

```python
# tampilkan pivot table dengan ROMI kumulatif yang telah dihitung
output.cumsum(axis=1).round(2).fillna('')
```

```python
# Menggabungkan dengan jumlah pelanggan unik per source_id
costs_per_source = pd.merge(total_costs_per_source, uid_source, on='source_id')

# Menampilkan hasil
print(costs_per_source[['source_id', 'costs', 'visitor']])
```

Selanjutnya Menghitung pendapatan pelanggan berdasarkan source_id lalu menggabungkannya dengan cost_per_souce

```python
#Gabungkan data_visits dan data_orders berdasarkan kolom 'uid':
merged_data = data_visits.merge(data_orders[['uid', 'revenue']], on='uid', how='left')
merged_data
```

```python
#Hitung total pendapatan (revenue) berdasarkan sumber (source_id):
revenue_per_source = merged_data.groupby('source_id')['revenue'].sum()
revenue_per_source.columns=['rev']
revenue_per_source
```

```python
#Gabungkan data costs_per_source dan profit_per_source berdasarkan kolom 'source_id':
merged_data = merged_data2.merge(profit_per_source, on='source_id')

merged_data.rename(columns={'revenue_y': 'laba_awal'}, inplace=True)
merged_data.rename(columns={'revenue_x': 'pendapatan'}, inplace=True)

#menampilkan hasil
merged_data
```

CAC (Customer Acquisition Cost):


```python
#CAC dapat dihitung dengan membagi total biaya (costs) dengan jumlah pelanggan baru (uid) yang diakuisisi dari masing-masing sumber. Misalnya:

merged_data['CAC'] = merged_data['costs'] / merged_data['visitor']
merged_data
```


#### 3. Seberapa menguntungkannya investasi tersebut? (ROI)

```python
#ROI (Return on Investment):
#ROI dapat dihitung dengan membagi total pendapatan (revenue) dengan total biaya (costs). Misalnya:


merged_data['ROI'] = merged_data['pendapatan'] / merged_data['costs']
merged_data
```

```python
#LTV (Lifetime Value):
#LTV dapat dihitung dengan membagi total pendapatan dengan jumlah pelanggan baru (visitor) yang diakuisisi dari masing-masing sumber


merged_data['LTV'] = merged_data['pendapatan'] / merged_data['visitor']
merged_data
```

Buatlah grafik yang menampilkan perbedaan metrik-metrik ini untuk berbagai perangkat dan sumber iklan, serta perubahannya dari waktu ke waktu.

```python
import numpy as np
import matplotlib.pyplot as plt

# Mengambil data yang diperlukan
source_ids = merged_data['source_id']
cac = merged_data['CAC']
roi = merged_data['ROI']
ltv = merged_data['LTV']

# Membuat indeks untuk posisi x pada plot
x = np.arange(len(source_ids))

# Menentukan lebar setiap bar
bar_width = 0.2

# Membuat plot
fig, ax = plt.subplots(figsize=(12, 8))
bar1 = ax.bar(x, cac, width=bar_width, label='CAC')
bar2 = ax.bar(x + bar_width, roi, width=bar_width, label='ROI')
bar3 = ax.bar(x + 2*bar_width, ltv, width=bar_width, label='LTV')

# Menambahkan nilai di atas setiap bar
def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{:.2f}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom', rotation=45)

autolabel(bar1)
autolabel(bar2)
autolabel(bar3)

# Mengatur label pada sumbu x dengan rotasi 45 derajat
ax.set_xticks(x + bar_width)
ax.set_xticklabels(source_ids, rotation=45)

# Mengatur label dan judul grafik
ax.set_xlabel('Source ID')
ax.set_ylabel('Value')
ax.set_title('Metrics by Source ID')

# Menampilkan legenda
ax.legend()

# Menampilkan grafik
plt.show()
```
