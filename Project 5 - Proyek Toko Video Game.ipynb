Sebenarnya saya kurang mengerti arti 'platform baru untuk muncul' namun dapat saya artikan menghitung waktu yang dibutuhkan platform baru untuk terjadi pertama kali penjualan. mungkin kurang lebih seperti itu.

```python
# hitung waktu yang dibutuhkan platform baru untuk terjadi penjualan

# Membuat DataFrame kosong untuk menyimpan hasil
hasil = pd.DataFrame(columns=['Platform', 'Waktu Penjualan (Tahun)'])

# Loop melalui setiap platform
for platform in platforms:
    # Filter data penjualan untuk platform yang dipilih
    penjualan_platform = data_games[data_games['platform'] == platform]

    # Dapatkan tahun rilis pertama platform
    tahun_rilis_pertama = penjualan_platform['year_of_release'].min()

    # Dapatkan tahun pertama terjadi penjualan di platform
    tahun_pertama_penjualan = penjualan_platform[penjualan_platform['total_sales'] > 0]['year_of_release'].min()

    # Hitung selisih tahun antara tahun rilis pertama dan tahun pertama penjualan
    waktu_penjualan = tahun_pertama_penjualan - tahun_rilis_pertama + 1

    # Tambahkan data ke DataFrame hasil
    hasil = hasil.append({'Platform': platform, 'Waktu Penjualan (Tahun)': waktu_penjualan}, ignore_index=True)

# Menampilkan hasil dalam bentuk tabel
print(hasil)
```

```python
import scipy.stats as stats

# Uji hipotesis untuk platform Xbox One dan PC
xone_ratings = data_filtered[data_filtered['platform'] == 'XOne']['user_score']
pc_ratings = data_filtered[data_filtered['platform'] == 'PC']['user_score']

t_statistic, p_value = stats.ttest_ind(xone_ratings, pc_ratings)
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

# Membandingkan p-value dengan tingkat signifikansi alpha
alpha = 0.05  # Ambang batas alpha
if p_value < alpha:
    print("Hipotesis nol ditolak. Terdapat perbedaan yang signifikan antara rating pengguna platform Xbox One dan PC.")
else:
    print("Hipotesis nol diterima. Tidak terdapat perbedaan yang signifikan antara rating pengguna platform Xbox One dan PC.")

```

Berdasarkan hasil hipotesis, Rata-rata rating pengguna platform Xbox One dan PC sama adalah benar. Hasilnya Hipotesis Tidak terdapat perbedaan yang signifikan antara rating pengguna platform Xbox One dan PC.


### Uji Hipotesis Rata-Rata Rating Pengguna Genre Action Dan Sports

— Rata-rata rating pengguna genre Action dan Sports berbeda. Tetapkan sendiri nilai ambang batas alpha.

```python

# Uji hipotesis untuk genre Action dan Sports
action_ratings = data_filtered[data_filtered['genre'] == 'Action']['user_score']
sports_ratings = data_filtered[data_filtered['genre'] == 'Sports']['user_score']

t_statistic, p_value = stats.ttest_ind(action_ratings, sports_ratings)
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

# Membandingkan p-value dengan tingkat signifikansi alpha
alpha = 0.05  # Ambang batas alpha

if p_value < alpha :
    print("Hipotesis nol ditolak. Terdapat perbedaan yang signifikan antara rating pengguna genre Action dan Sports.")
else:
    print("Hipotesis nol diterima. Tidak terdapat perbedaan yang signifikan antara rating pengguna genre Action dan Sports.")
```

Berdasarkan hasil hipotesis, Rata-rata rating pengguna genre Action dan Sports berbeda adalah benar. hasilnya Terdapat perbedaan yang signifikan antara rating pengguna genre Action dan Sports.
